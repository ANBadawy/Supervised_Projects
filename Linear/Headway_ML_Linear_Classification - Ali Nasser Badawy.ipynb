#### To find the equation of the decision boundary line, we will use the centroids we calculated earlier and the middle points using the given formula above.

#### $\frac{x_1-x'_1} {x''_1-x'_1} = \frac{x_2-x'_2} {x''_2-x'_2} $ will help in obtaining the equation of the line connecting the centroids.

#### This line connecting the centroids will cut the middle point we calculated, the equation of the decision boundary formed by creating a perpendicular line on the centroids connecting line and middle point.

#### After providing the concept; now, let's calculate the centroids connecting line.

#### $\frac{x_1-7}{3-7} = \frac{x_2-4}{8-4}$, for simplicity the equation will be $x_2 = -x_1 + 11$
#### knowing that the slope equation formula is $x_2 = a.x_1 + b$. The decision boundary is negative reciprocal of $a$, which is $-1$, then $a = 1$ equation: $x_2 = x_1 + b$
#### then using the middle point, $b$ is calculated. middle point = $[5, 6]$, by substuting in the equation above: $6 = 5 + b$, resulting in $b = 1$

```python
# First Perceptron code implemented
def perceptron(data, w=[0, 0], learning_rate=1):

    bias, converged = 0, False

    while not converged:

        converged = True
        
        for row in data.itertuples():

            x_i, y_i = np.array([row.x1, row.x2]), row.y
            
            if y_i * (np.dot(w, x_i) + bias) <= 0:

                w += learning_rate * y_i * x_i
                bias += learning_rate * y_i
                converged = False
    
    return w, bias

perceptron(data)
```

```python
def decision_boundary_plot(data, weights, bias):

    x1_values = np.linspace(0, 8, 100)

    x2_values = -1 * ((weights[0] / weights[1]) * x1_values) - (bias / weights[1])

    
    plt.plot(data[data['y'] == 1]['x1'], data[data['y'] == 1]['x2'], 'bo', label='Class 1' )
    plt.plot(data[data['y'] == -1]['x1'], data[data['y'] == -1]['x2'], 'ro', label='Class -1')
    
    plt.plot(x1_values, x2_values, color='green', label='Decision Boundary', linestyle='--')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title('Perceptron Decision Boundary')
    plt.legend()
    plt.show()

# Unpack
weights, bias = perceptron(data)

# Plot the decision boundary
decision_boundary_plot(data, weights, bias)
```

```python
def dual_perceptron(data):
    ##### YOUR CODE STARTS #####(please do not delete this line)
    
    learning_rate, samples_num, converged = 1, len(data), False
    alpha, m_matrix = np.zeros(samples_num), np.zeros((samples_num, samples_num))

    # calculating dot product for the matrix
    for i in range(samples_num):
        for j in range(samples_num):
            x_i = np.array([data.iloc[i]['x0'], data.iloc[i]['x1'], data.iloc[i]['x2'] ])
            x_j = np.array([data.iloc[j]['x0'], data.iloc[j]['x1'], data.iloc[j]['x2'] ])
            m_matrix[i, j] = np.dot(x_i, x_j)
    
    while not converged:
        converged = True

        for index, row in data.iterrows():

            alpha_sum, y_i = np.sum([alpha * data['y'] * m_matrix[index, 0: ]]), row['y']
            
            # if alpha_sum * y_i > 0 then it not misclassified else it is
            if alpha_sum * y_i <= 0:
                converged = False
                alpha[index] += learning_rate

    # calculates the w by forming alpha as a column and data['y'] too
    w = np.sum(alpha.reshape(-1, 1) * data['y'].values.reshape(-1, 1) * data[['x0', 'x1', 'x2']].values, axis=0)

    return w

    raise NotImplementedError("Compare different methods")
    ##### YOUR CODE ENDS #####(please do not delete this line)
    
weights = dual_perceptron(data)
print("Weight vector w:", weights)
```
